```python
04
class Solution:
    def findNumberIn2DArray(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix:
            return False
        row = len(matrix)
        col = len(matrix[0])
        i,j = 0,col - 1
        while i < row and j >= 0:
            if matrix[i][j] > target:
                j -= 1
            elif matrix[i][j] < target:
                i += 1
            else:
                return True
        return False
```

```python
10.2 DP
class Solution:
    def numWays(self, n: int) -> int:
        DP = [0] * (n+1)
        if n == 0:
            DP[0] = 1
        else:
            DP[0] = 1
            DP[1] = 1
        for i in range(2,n+1):
            DP[i] = DP[i-1] + DP[i-2]
        return DP[n]%1000000007     
```

```python
13.
#A1 BFS  O(MN)
class Solution:
    def movingCount(self, m: int, n: int, k: int) -> int:
        visited = [[1 for i in range(n)] for j in range(m)]
        for i in range(m):
            for j in range(n):
                if i // 10 + i % 10 + j // 10 + j % 10 <= k:
                    visited[i][j] = 0
        if k < 0:
            return -1
        q = collections.deque()
        q.append((0,0))
        res = 1
        visited[0][0] = 1
        while q:
            x, y = q.popleft()
            direction = [(0,1),(1,0)]
            for delta_x, delta_y in direction:
                next_x = x + delta_x
                next_y = y + delta_y
                if self.can_visit(next_x,next_y,visited,m,n):
                    q.append((next_x, next_y))
                    res += 1
                    visited[next_x][next_y] = 1
        return res
    
    def can_visit(self,next_x, next_y,visited,m,n):
        if next_x < 0 or next_x >= m:
            return False
        if next_y < 0 or next_y >= n:
            return False
        if visited[next_x][next_y] == 1:
            return False
        return True

#A2 DFS  
```

```python
14.1 DP problem
class Solution:
    def cuttingRope(self, n: int) -> int:
        memo = [-1] * (n + 1)
        for i in range(2, n + 1):
            for j in range(1, i):
                memo[i] = max(memo[i], j*memo[i-j], j * (i-j))
        return memo[n]
```

```python
# 24 reverse the linked-list 
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prev, curr = None, head
        while curr:
            # curr.next,prev,curr = prev,curr,curr.next
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev
```
